```python
gpu_df = pd.read_csv('All_GPUs.csv')
gpu_df.head()
gpu_df['Release_Date'] = gpu_df['Release_Date'].str[1:-1]
```

```python
gpu_df.head()
gpu_df=gpu_df[gpu_df['Release_Date'].str.len()==11]
gpu_df.head()
gpu_df['Release_Date']=pd.to_datetime(gpu_df['Release_Date'], format = '%d-%b-%Y')
```

```python
gpu_df.head()
```

```python
gpu_df.info
```

```python
gpu_df['Release_Year']=gpu_df['Release_Date'].dt.year
gpu_df['Release_Quarter']=gpu_df['Release_Date'].dt.quarter
gpu_df['Release_Month']=gpu_df['Release_Date'].dt.month
```

```python
gpu_df.head()
```

```python
plt.figure(figsize=(13, 6))
gpu_df['Release_Month'].groupby(gpu_df['Release_Month']).count().plot(kind='bar')
plt.title('Resolution counts')
plt.xlabel('Release Month')
plt.ylabel('Count of GPU releases')
plt.show()
```
